```python
df1.columns = ['id','title','cast','crew']
df3= df2.merge(df1,on=['id','title'])
print(df3.shape)
print(df3.info())
df3.head(3)
```

Lets replace the missing Values of Homepage, release_date and runtime with an Empty string

```python
df3['homepage'] = df3['homepage'].fillna(' ')
df3['release_date'] = df3['release_date'].fillna(' ')
df3['runtime'] = df3['runtime'].fillna(' ')
df3['tagline'] = df3['tagline'].fillna(' ')
```

As we can see the merging worked perfectly so we have 22 columns instead of 24 columns.

We can also observe from the table that some of the columns are in json format,made of a list of dictionaries.

this list of dictionaries contain many values some of which are not important(like id) as these information are already represented by the respective values, so lets define some functions that would help extract the keywords and append them into a list. the target columns are (genres,keywords,production_companies,spoken_languages,cast and crew). But we would define a separate function for cast and crew as they contain more than 2 important attributes(like the actor's name and sex or position for crew members).

Here i applied the literal_eval function to the target columns of the dataframe.

The literal_eval safely evaluates an expression, node or a string containing a Python
expression

```python
from ast import literal_eval
col_name = ['genres','keywords','production_companies','spoken_languages','cast','crew']
for cols in col_name:
    df3[cols] = df3[cols].apply(literal_eval)
```

This function below creates a list of the important features from the Dataframe 

```python
def get_features(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        
        return names

    #Return empty list in case of missing Data
    return []
```

```python
#notice that 'crew' are 'cast' are excluded from the iteration here.
col_names = ['genres','keywords','production_companies','spoken_languages']
for cols in col_names:
    df3[cols] = df3[cols].apply(get_features)
```

Now let us focus on the Crew and cast columns. unlike the other columns the crew and cast columns have more than one important attributes embedded in the list of dictionaries(json), attributes like:

1. name : ....

Now lets write a function that extracts the job, name and gender as dictionary in the crew column

```python
#function for crew memebers
keys = ['gender','job','name']
def get_crew_role(x):
    if isinstance(x, list):
        names = {i['name']: {'gender':i['gender'],'job': i['job']} for i in x}
        
        return names

    #Return empty list in case of missing Data
    return []
df3['crew'] = df3['crew'].apply(get_crew_role)
```

```python
#function for cast actors
keys = ['name','gender','order']
def  get_cast_role(x):
    if isinstance(x,list):
        roles = {i['name']: {'gender': i['gender'], 'order':i['order']} for i in x}
        
        return roles
    
    return []
df3['cast'] = df3['cast'].apply(get_cast_role)
```
